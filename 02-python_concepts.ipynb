#### 7. Write a short program that displays the future value of 1,000 earning 5%  at the end of each year for the next 5 years

```python
# Prof's ans
pv = 100
rate =.05
print(f"{'Year':<15}{'Balance'}")
for year in range(5):
    print({f"{year:<15d}{pv:>,.0f}"})
    pv *= (1 + rate)
```

```python
def future_value(present_value, annual_rate, years):
    future_value = present_value * (1 + annual_rate) ** years
    return future_value

present_value = 1000
annual_rate = 0.05

for year in range(1, 6):
    value = future_value(present_value, annual_rate, year)
    print(f"Year {year}: ${value:.2f}")
```

#### 9. Write the code necessary to calculate the sum of the square differences of a group of values : 
58, 32, 37, 41, 36, 36, 54, 37, 25, 53
##### $\Sigma $(${x}$ - $\bar{x}$)$^2$

```python
values = [58, 32, 37, 41, 36, 36, 54, 37, 25, 53]

def calculate_sum_of_squares_difference(values):
    mean = sum(values) / len(values)
    sum_of_squares_difference = 0
    
    for x in values:
        sum_of_squares_difference += (x - mean) ** 2
    
    return sum_of_squares_difference

result = calculate_sum_of_squares_difference(values)
result
```

#### 12. Write the code needed to evaluate whether a value in a group of values is negative or postive and prints an approriate message, i.e. postive, negative 
-1.59, 2.36, 1.69, 1.13, -0.91, 1.48, -0.34, 1.31, -0.74, 0.2

```python
values = [-1.59, 2.36, 1.69, 1.13, -0.91, 1.48, -0.34, 1.31, -0.74, 0.2]

for value in values:
    if value > 0:
        print(f"{value} is positive")
    elif value < 0:
        print(f"{value} is negative")
    else:
        print(f"{value} is zero")
```

#### 14.  Complete the following:
1. Write a program that generates 500 random integers from 1 to n, for example, the first integer will be 1 the second will be 1 or 2, the fifth between 1 and 5, and so on.
2. The program should write each integer to a file 
3. Once all integers are written close the file

##### Part 2
1. Open and read the file into a list
2. Create a line plot of the data

```python
# Part 1
import random

# Open file for writing
with open("random_numbers.txt", "w") as file:
    for n in range(1, 501):
        number = random.randint(1, n)
        file.write(str(number) + "\n")
```

```python
# Part 2
import matplotlib.pyplot as plt

# Read the file into a list
numbers = []

with open("random_numbers.txt", "r") as file:
    for line in file:
        numbers.append(int(line.strip()))

# Create a line plot
plt.plot(numbers)
plt.xlabel("Index")
plt.ylabel("Value")
plt.title("Line Plot of Random Integers")
plt.show()
```
